```python
train_data=train[:int(0.8*len(train))]
val_data=train[int(0.8*len(train)):]
```

```python
train_data.describe()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

#Define features and targets
features=list("ABCDEFGHIJKLMN")
target_cols=['Y1','Y2']


```

```python
#Preparing train and validation features and targets
X_train=train_data[features]
y_train=train_data[target_cols]
X_val=val_data[features]
y_val=val_data[target_cols]

```

```python
#Initialize a separate model for each target (or multi-output model)
model_y1=RandomForestRegressor(random_state=42)
model_y2=RandomForestRegressor(random_state=42)

```

```python
#Train models
model_y1.fit(X_train,y_train['Y1'])
```

```python
#Predict on validation data
y1_pred=model_y1.predict(X_val)
```

```python
r2_y1=r2_score(y_val['Y1'],y1_pred)
```

```python
print(f"Validation R2 for-Y1: {r2_y1:.4f}")
```

```python
model_y2.fit(X_train,y_train['Y2'])

```

```python



#Predict on validation data
y2_pred=model_y2.predict(X_val)

```

```python
#Calculate R2 scores

r2_y2=r2_score(y_val['Y2'],y2_pred)
avg_r2=(r2_y1 + r2_y2)/2


print(f"Validation R2 for-Y2: {r2_y2:.4f}")
print(f"Average Validation-R2: {avg_r2:.4f}")
```

```python
#features=["A","B",...,"N"] as previously defined
#For Y1 model
importances_y1=model_y1.feature_importances_

importance_df_y1=pd.DataFrame({
    'feature': features,
    'importance': importances_y1
}).sort_values(by='importance',ascending=False)


print("Feature importances for Y1 model:")
print(importance_df_y1)


```

```python
#for Y2 model
importances_y2=model_y2.feature_importances_

importance_df_y2=pd.DataFrame({
    'feature': features,
    'importance': importances_y2
}).sort_values(by='importance',ascending=False)

print("\nFeature importances for-Y2 model:")
print(importance_df_y2)
```
